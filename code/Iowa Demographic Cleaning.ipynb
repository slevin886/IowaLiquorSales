```python
import pandas as pd
df = pd.ExcelFile('../IPE01.xls')
xls = pd.read_excel('../Mastdata.xls')
```

```python
df1 = pd.read_excel(df, 'Sheet1')

def colfinder(df):
    t = []
    for col in df.columns:
        try: 
            p = xls.loc[xls['Item_Id'] == col]
            some = list(p['Item_Description'].values)[0]
            t.append(some)
        except:
            t.append(col)
    df.columns = t
    new_df = df[df['Areaname'].str.contains(', IA')]
    return new_df

demographics = colfinder(df1)[['Areaname', 'Median household income 2005']].reset_index(drop=True)
demographics
```

```python
df2 = pd.read_excel(df, 'Sheet2')
df2 = colfinder(df2)['Median household income 2009']
demographics['Median household income 2009'] = df2.values
df3 = pd.read_excel(df, 'Sheet3')
df3 = colfinder(df3)['People of all ages in poverty - number 2009'].values
demographics['All People in poverty- 2009'] = df3
#Finished with this excel sheet. 


```
